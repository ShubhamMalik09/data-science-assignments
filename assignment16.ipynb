```python
#Q1. Explain GET and POST methods.
sol="GET and POST are two commonly used HTTP methods in web development. They are used to send data from the client to the server and vice versa."

# GET method
sol=" In the GET method, data is appended to the URL as query parameters.\
Since the data is included in the URL, it is visible and can be bookmarked, shared, and easily accessed.\
The GET method is primarily used to retrieve data from the server. It should not be used to transmit sensitive information,\
such as passwords, as the data is visible in the URL"

# POST method
sol="In the POST method, data is sent in the request body, separate from the URL.\
The data is not visible in the URL, making it more secure for sensitive information.\
The POST method is used to submit data to the server, often used for form submissions, login credentials,\
or other sensitive data that should not be visible in the URL."
```

```python
#Q2. Why is request used in Flask?
sol="In Flask, the request object is used to access incoming data from the client.\
The request object is an instance of the Request class which is imported from the Flask module.\
The request object is essential for handling file uploads, accessing client information like IP addresses and user agents,\
and working with dynamic URLs and routing. It enables Flask web applications to interact with users, retrieve user input,\
and respond appropriately based on the data received in the client requests."
```

```python
#Q3. Why is redirect() used in Flask?
sol="In Flask, the redirect() function is used to perform a redirection from one URL to another.\
It is a way to send an HTTP redirect response back to the client's web browser, instructing it to navigate to a different URL.\
It is used in scenerios like when a user submits a form and the server needs to redirect them to another page after processing the form data."
```

```python
#Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.
from flask import Flask,request,jsonify

app = Flask(__name__)

@app.route("/",methods=['GET','POST'])
def hello_world():
    return "<h1>HELLO</h1>"

@app.route("/welcome",methods=['POST'])
def welcome():
    if(request.method=="POST"):
        first_name=request.json["first_name"]
        last_name=request.json["last_name"]
        result= "Welcome to our app {} {}".format(first_name,last_name)
        return jsonify(result)
    

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
