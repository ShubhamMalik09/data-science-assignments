```python
#Q3. What is App routing in Flask? Why do we use app routes?
sol="In Flask, app routing refers to the process of defining URL patterns (routes) and associating them \
with specific viewfunctions in a web application. It allows developers to determine how different URLs should be handled\
and what content or response should be displayed when a client makes a request to a particular URL. \
\
App routes provide a way to map URLs to corresponding view functions. When a client makes a request to a specific URL,\
Flask uses the defined app routes to determine which view function should handle the request.\
For building RESTful APIs, app routes are crucial. Each route in the API corresponds to a specific endpoint that handles\
different HTTP methods (GET, POST, PUT, DELETE) and performs corresponding actions on the resources.\
With app routes, developers can use dynamic route patterns, allowing parts of the URL to be variable. \
This flexibility is useful for handling different types of requests and passing parameters to view functions."
```

```python
#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
sol="In Flask, the url_for() function is used for URL building. It generates a URL for a specific view function based\
on its name and arguments. Using url_for() ensures that the URLs are dynamically generated and avoids potential issues if routes change."

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the Home Page!"

@app.route('/about')
def about():
    return "This is the About Page!"

with app.test_request_context():
    home_url = url_for('home')
    about_url = url_for('about')
    print("Home Page URL:", home_url)
    print("About Page URL:", about_url)
```
